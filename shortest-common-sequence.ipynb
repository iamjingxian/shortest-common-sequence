```python

# initialise the two strings
p = ""
q = ""

# open file
with open("./Two_strings.txt","r") as file:
    # strip away the empty lines
    lines = (line.rstrip() for line in file)
    # create a list that contain all the non-blank lines
    lines = list(line for line in lines if line) # Non-blank lines in a list

# assign the two strings    
p = lines[0]
q = lines[1]

m = len(p) # letters in p are p[0], p[1],...p[m-1]
n = len(q)

# step 1: Compute the length of the SCS;

# initialise with an empty m  x n table
table = [[0 for i in range(len(p)+1)] for j in range(len(q)+1)]

# populate 1st row and 1st column with running index
for r in range(len(q)+1):
    table[r][0] = r
for c in range(len(p)+1):
    table[0][c] = c

# start from the "1st" row, ie from the 1st letter in q
for r in range(1, len(q)+1):
    #then, go down the column
    for c in range(1,len(p)+1):
        if q[r-1] == p[c-1]:
            table[r][c] = table[r-1][c-1]+1
        else:
            table[r][c] = min(table[r-1][c], table[r][c-1])+1

            
# step 2: Infer an SCS from the computation in Step 1.

# let l = length of scs, and initialize a string
scslength = table[n][m] 
scs = ""

# now, work "backwards" from the last letter(s) of scs

i = m # for letters in p
j = n # for letters in q

#stop when reach 1st letter of either p or q, ie p[0] or q[0] = 0
while i * j > 0: 
    
    # If current letter in p and q are same, 
    if p[i - 1] == q[j - 1]:
        # add letter in p to scs
        scs += p[i-1]
        # since this is a while loop; 
        # need to reduce i or j after each iteration, 
        # so that the loop stops eventually
        i -= 1
        j -= 1
 
    # Else, if current letters in p and q are diff
    # case 1 (of 2) value of cell above is > value of cell to the left
    elif table[j - 1][i] > table[j][i - 1]:
        # add letter in p to scs
        scs += p[i - 1]
        # reduce values of j and index
        i -= 1
    
    # If current letters in p and q are diff
    # case 2 (of 2) value of above is < value of cell to the left
    else:
 
        # add letter in q to scs
        scs += q[j - 1]
 
        # reduce values of i and index
        j -= 1

# If q has reached its end 
# => j = 0 => i*j == 0 
# but i != 0, means 
while i > 0:
    # add remaining letters in p to scs
    scs += p[i - 1]
    i -= 1

# If p has reached its end
# => i = 0 => i*j == 0 but j != 0
while j > 0:
    # add remaining letters in q to scs
    scs += q[j - 1]
    j -= 1

scs = scs[::-1]
print(scs)

```
